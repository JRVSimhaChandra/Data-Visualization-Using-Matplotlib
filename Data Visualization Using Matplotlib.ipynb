```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'C':20, 'C++':15, 'Java':30,
        'python':50}
courses = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10, 5))

plt.bar(courses, values, color ='green')

plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

# Box Plot

Detecting the outliers

```python
data=[np.random.normal(20,std,165) for std in range(1,4)]
data
```

```python
x=np.random.normal(20,1,165)
y=np.random.normal(20,2,165)
z=np.random.normal(20,3,165)
data=[x,y,z]
```

```python
plt.boxplot(data)
plt.show()
```

# Pie Plot
 Use it for categorical and discrete 

```python
# pie()
pielist=[1,2,3,4,8]
plt.pie(pielist,labels=['Mango','Apple','banana','Pine','Orange'])
plt.show()
```

```python
# task plot categorical variable in pie representation
```

```python
lab= ['Python', 'C++','Ruby','Java']
sizes= [215,130,245,210]
color=['gold','yellowgreen','lightcoral','lightskyblue']

plt.pie(sizes,labels=lab,colors=color)
plt.title('courses offered')
plt.show()
```

```python
lab= ['Python', 'C++','Ruby','Java']
sizes= [215,130,245,210]
color=['gold','yellowgreen','lightcoral','lightskyblue']
explode=[0,0.1,0,0]

plt.pie(sizes,labels=lab,colors=color,autopct='%.3f%%',explode=explode)
plt.title('courses offered')
plt.show()
```
